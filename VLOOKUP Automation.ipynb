### Function to Extract Files

```python
def extract_files(file1,file2,sheet1,sheet2):
    df1=pd.read_excel(file1, sheet_name=sheet1,    )
    df2=pd.read_excel(file2, sheet_name=sheet2)
    
    return (df1,df2)
```

### Function to Identify the Columns Name

```python

def select(com):
    man=[]
    common=[]
    if com.lower()=="manual":
        no=int(input("Enter the number of common columns you want to apply VLOOKUP:  "))
        i=0
        while i<no:
            name=input("Enter the common column name in exact format given in table: ")
            man.append(name)
            i=i+1
            
    elif com.lower()=="common":
        for i in list(df1.columns):
            if i in list(df2.columns):
                common.append(i)

    else:
        print("Kindly try again !!")
    
    return(man,common)
```

#### Function to Proceed with Common Columns

```python

def add(apply):
    i=0
    cols=[]
    while i<apply:
        col=input("Enter column " + str(i) + " name in exact format given in table 2 : ")
        cols.append(col)
        i=i+1
    return (cols)
```

### Alotting the values based on common columns 

```python
def vlookup(cols,common):
    for match_col in cols:
        df1[match_col]=" "
        for j in df1.index:
            for k in df2.index:
                if (df1[common].loc[j]==df2[common].loc[k]).all(axis=0) == True:
                    df1[match_col][j]=df2[match_col][k] 
                    
    return (df1,df2)
```

```python
def final(cols,man,common):
    if len(man)==0:
        df1,df2=vlookup(cols,common)
    
    else:
        df1,df2=vlookup(cols,man)
        
    return (df1,df2)
```
